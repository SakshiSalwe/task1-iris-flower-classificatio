```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.barplot(x = 'Class', y = 'Sepal length', data = data1)
plt.subplot(2,2,2)
sns.barplot(x = 'Class', y = 'Petal length', data = data1)
plt.subplot(2,2,3)
sns.barplot(x = 'Class', y = 'Sepal width', data = data1)
plt.subplot(2,2,4)
sns.barplot(x = 'Class', y = 'Petal width', data = data1)
```

```python
setosa = data1[data1['Class'] == "Iris-setosa"]
versicolor = data1[data1['Class'] == "Iris-versicolor"]
virginica = data1[data1['Class'] == "Iris-virginica"]

plt.figure(figsize=(10,10))
plt.scatter(setosa['Sepal length'],setosa['Sepal width'], c="blue", label="Iris setosa",marker = '*')
plt.scatter(setosa['Sepal length'],versicolor['Sepal width'], c="red", label="versicolor",marker = '^')
plt.scatter(setosa['Sepal length'],virginica['Sepal width'], c="green", label="virginica",marker = '<')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Sepal length vs Sepal width", fontsize=16)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(setosa['Petal length'],setosa['Petal width'],c="blue",label="Iris-setosa",marker='*')
plt.scatter(versicolor['Petal length'],versicolor['Petal width'],c="red",label="versicolor",marker='^')
plt.scatter(virginica['Petal length'],virginica['Petal width'],c="green",label="virginica",marker='<')
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Petal length vs Petal width",fontsize=16)
plt.legend()
plt.show()
```

```python
x = data1.drop(columns="Class")
y = data1["Class"]
```

```python
plt.figure(figsize=(20,20),facecolor='white')
plotnumber = 1
for column in x:
  if plotnumber<=9 :
    ax = plt.subplot(3,3,plotnumber)
    sns.stripplot(y,x[column])
    plotnumber+=1
    plt.tight_layout()
```
